### 1. Apakah yang dimaksud dengan `coeficient` pada `Logistic Regression`?

- Jawaban
Dalam Logistic Regression, "coefficient" (koefisien) mengacu pada nilai yang mengukur tingkat pengaruh atau bobot setiap variabel independen terhadap log-odds dari variabel target. Koefisien ini menunjukkan seberapa besar perubahan dalam log-odds target yang diharapkan terjadi jika variabel independen mengalami perubahan satu satuan. Dengan kata lain, koefisien menggambarkan sejauh mana variabel independen berkontribusi terhadap prediksi probabilitas dalam model Logistic Regression.

### 2. Apakah fungsi parameter kernel pada `SVM`? Jelaskan salah satu kernel yang kalian pahami!

- Jawaban <br>
Parameter kernel pada SVM (Support Vector Machine) adalah parameter yang menentukan jenis fungsi kernel yang digunakan untuk mengubah data ke dalam ruang dimensi yang lebih tinggi. Kernel digunakan untuk mengukur kesamaan antara dua titik data dalam ruang karakteristik yang tinggi. Beberapa kernel yang umum digunakan dalam SVM adalah:

1. **Linear Kernel**: Kernel ini adalah kernel paling sederhana dan digunakan untuk masalah klasifikasi linear. Ini mengukur kesamaan antara dua titik sebagai produk titik antara vektor fitur. Kernel ini cocok untuk masalah di mana batas keputusan adalah linier.

2. **RBF (Radial Basis Function) Kernel**: Kernel RBF adalah salah satu yang paling umum digunakan dalam SVM. Ini mentransformasikan data ke dalam ruang dimensi yang tak terhingga. Kernel RBF dapat menangani masalah non-linier dan memungkinkan pembuatan batas keputusan yang lebih kompleks.

3. **Polynomial Kernel**: Kernel ini digunakan untuk masalah yang memiliki pola non-linier dengan tingkat tertentu. Parameter derajat kernel polinomial menentukan sejauh mana data diubah menjadi ruang polinomial. Kernel ini cocok untuk masalah yang memiliki pola polinomial.

Jadi, parameter kernel memungkinkan untuk memilih fungsi kernel yang sesuai dengan karakteristik data, yang mempengaruhi kemampuan SVM untuk menangani masalah klasifikasi dengan berbagai tingkat kompleksitas.

### 3. Bagaimana cara memilih `K` yang optimal pada `KNN` ?

- Jawaban <br>

Cara memilih nilai `K` yang optimal pada K-Nearest Neighbors (KNN) dengan bantuan `RandomSearchCV` melibatkan proses yang dikenal sebagai "cross-validation" untuk mengevaluasi kinerja model KNN dengan berbagai nilai `K` seperti yang dilakukan pada GC 5 ini semua model menggunakan `RandomSearchCV`. Berikut adalah langkah-langkahnya:

1. **Pilih Rentang Nilai K**: Tentukan rentang nilai yang akan eksplorasi untuk `K`. Misalnya, bisa memilih nilai dari 1 hingga sejumlah tetangga yang mungkin, atau bisa menggunakan metode trial and error.

2. **Validasi Silang (Cross-Validation)**: Gunakan teknik validasi silang, seperti validasi silang K-fold, untuk membagi data menjadi set pelatihan dan set validasi. Misalnya, dapat membagi data menjadi 5 atau 10 lipatan (folds). Ini membantu menghindari overfitting dan menghasilkan perkiraan kinerja yang lebih stabil.

3. **Hitung Kinerja Model**: Untuk setiap nilai `K` dalam rentang yang pilih, latih model KNN dengan nilai `K` tersebut pada set pelatihan dan evaluasi pada set validasi. Hitung metrik evaluasi seperti akurasi, presisi, recall, F1-score, atau yang sesuai dengan kasus.

4. **Pilih Nilai K Terbaik**: Pilih nilai `K` yang memberikan kinerja terbaik berdasarkan metrik evaluasi yang tentukan. Ini dapat dilakukan dengan membandingkan metrik evaluasi antara berbagai nilai `K` yang dievaluasi.

5. **Validasi Hasil**: Setelah memilih nilai `K` yang optimal, perlu diuji ulang pada set data uji yang independen yang tidak digunakan selama proses penyesuaian.

6. **Pemantauan Model**: Ingatlah bahwa pemilihan nilai `K` yang optimal dapat berubah seiring dengan perubahan data atau perubahan dalam tugas klasifikasi. Oleh karena itu, model perlu dipantau secara berkala dan diuji ulang dengan nilai `K` yang optimal.

Pemilihan nilai `K` yang optimal melalui proses cross-validation membantu memastikan bahwa model KNN memiliki kinerja yang baik dan tidak overfitting atau underfitting terhadap data.

### 4. Apa yang dimaksud dengan metrics-metrics berikut : `Accuracy`, `Precision`, `Recall`, `F1 Score`, dan kapan waktu yang tepat untuk menggunakannya ?


1. **Accuracy (Akurasi)**:
   - **Definisi**: Akurasi adalah metrik yang mengukur sejauh mana model klasifikasi benar dalam memprediksi label secara keseluruhan.
   - **Kapan Menggunakan**: Akurasi cocok digunakan ketika kelas-kelas dalam dataset seimbang dan memiliki frekuensi yang hampir sama. Ini cocok untuk masalah klasifikasi dengan tujuan mengklasifikasikan sebanyak mungkin dengan benar.

2. **Precision (Presisi)**:
   - **Definisi**: Presisi adalah metrik yang mengukur sejauh mana prediksi positif model adalah benar. Ini mengukur tingkat keakuratan positif dari semua prediksi positif.
   - **Kapan Menggunakan**: Presisi cocok digunakan ketika salah positif memiliki dampak yang tinggi atau ketika ingin memastikan bahwa ketika model memprediksi positif, itu benar.

3. **Recall (Recall atau Sensitivitas)**:
   - **Definisi**: Recall adalah metrik yang mengukur sejauh mana model mampu mengidentifikasi semua instance positif yang sebenarnya. Ini mengukur tingkat keakuratan positif dari semua instance positif.
   - **Kapan Menggunakan**: Recall cocok digunakan ketika salah negatif memiliki dampak yang tinggi atau ketika ingin memastikan bahwa model dapat mengidentifikasi sebanyak mungkin instance positif.

4. **F1 Score**:
   - **Definisi**: F1 Score adalah metrik yang menggabungkan presisi dan recall menjadi satu metrik tunggal. Ini adalah rata-rata harmonik dari keduanya.
   - **Kapan Menggunakan**: F1 Score cocok digunakan ketika ingin mencari keseimbangan antara presisi dan recall. Ini berguna ketika keseimbangan antara salah positif dan salah negatif penting.

Waktu yang tepat untuk menggunakan metrik-metrik ini tergantung pada konteks masalah dan prioritas bisnis. Misalnya, dalam kasus klasifikasi spam email, mungkin lebih peduli dengan presisi untuk meminimalkan email yang sah yang disaring sebagai spam (salah positif), sementara dalam kasus deteksi penyakit, mungkin lebih peduli dengan recall untuk memastikan bahwa tidak ada kasus penyakit yang terlewatkan (salah negatif). Akurasi cocok digunakan ketika ingin mengevaluasi kinerja secara keseluruhan, tetapi hanya jika kelas dalam dataset seimbang.

````python
# Variabel target dan variabel numerik
target = df['default_payment_next_month']
numeric_features = ['limit_balance', 'age', 'bill_amt_1', 'bill_amt_2', 'bill_amt_3', 'bill_amt_4', 'bill_amt_5', 'bill_amt_6',
                   'pay_amt_1', 'pay_amt_2', 'pay_amt_3', 'pay_amt_4', 'pay_amt_5', 'pay_amt_6']

# Menyiapkan list untuk hasil korelasi dan label variabel
correlations = []
labels = []

# Menghitung korelasi dengan pendekatan point biserial
for feature in numeric_features:
    point_biserial_correlation, p_value = stats.pointbiserialr(target, df_numeric[feature])
    correlations.append(point_biserial_correlation)
    labels.append(feature)

# Membuat plot
plt.figure(figsize=(10, 6))
plt.barh(labels, correlations, color='skyblue')
plt.xlabel('Korelasi Point Biserial')
plt.title('Korelasi Point Biserial antara default_payment_next_month dan Variabel Numerik')
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Menampilkan plot
plt.show()
````

### 2. Korelasi Binary dengan Ordinal

````python
# Daftar variabel ordinal
ordinal_variables = ['education_level', 'pay_0', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']
df_ordinal = df[ordinal_variables]

# Hitung korelasi Spearman dan buat plot
for var in ordinal_variables:
    corr, p_value = stats.spearmanr(df['default_payment_next_month'], df_ordinal[var])
    
    plt.figure(figsize=(6, 4))
    plt.scatter(df['default_payment_next_month'], df_ordinal[var], alpha=0.5)
    plt.xlabel('default_payment_next_month')
    plt.ylabel(var)
    plt.title(f'Spearman Correlation between default_payment_next_month and {var}\nCorrelation: {corr:.2f}, p-value: {p_value:.4f}')
    plt.show()
````

### 3. Korelasi Binary dengan Categorical (Nominal)

````python
from scipy.stats import chi2_contingency

# Membuat tabel kontingensi untuk variabel 'sex' dan 'default_payment_next_month'
contingency_sex = pd.crosstab(df['sex'], df['default_payment_next_month'])

# Membuat tabel kontingensi untuk variabel 'marital_status' dan 'default_payment_next_month'
contingency_marital = pd.crosstab(df['marital_status'], df['default_payment_next_month'])

# Melakukan uji chi-square untuk 'sex'
chi2_sex, p_sex, _, _ = chi2_contingency(contingency_sex)

# Melakukan uji chi-square untuk 'marital_status'
chi2_marital, p_marital, _, _ = chi2_contingency(contingency_marital)

# Menghitung koefisien phi (φ) untuk 'sex'
n_sex = contingency_sex.values.sum()
phi_corr_sex = np.sqrt(chi2_sex / n_sex)

# Menghitung koefisien phi (φ) untuk 'marital_status'
n_marital = contingency_marital.values.sum()
phi_corr_marital = np.sqrt(chi2_marital / n_marital)

# Menampilkan hasil uji chi-square, p-value, dan koefisien phi (φ)
print("Hasil uji chi-square untuk 'sex':")
print(f"Chi-square value: {chi2_sex}")
print(f"P-value: {p_sex}")
print(f"Phi correlation: {phi_corr_sex:.4f}")

print("\nHasil uji chi-square untuk 'marital_status':")
print(f"Chi-square value: {chi2_marital}")
print(f"P-value: {p_marital}")
print(f"Phi correlation: {phi_corr_marital:.4f}")

# Plot tabel kontingensi untuk 'sex'
sns.heatmap(contingency_sex, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Tabel Kontingensi 'sex' vs 'default_payment_next_month'")
plt.show()

# Plot tabel kontingensi untuk 'marital_status'
sns.heatmap(contingency_marital, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Tabel Kontingensi 'marital_status' vs 'default_payment_next_month'")
plt.show()
````

# VI. Model Definition

## Definisi Model


Model yang digunakan yaitu `Logistic Regression, SVM, dan KNN.`, Alasan penggunaannya tiap model yaitu:

1. **Logistic Regression**:
   - **Alasan**: Logistic Regression adalah pilihan umum ketika memiliki masalah klasifikasi biner. Dalam kasus ini, target variabel adalah 'default_payment_next_month' yang merupakan masalah klasifikasi biner (0 untuk tidak gagal bayar, 1 untuk gagal bayar).
   - **Keuntungan**: Model ini relatif sederhana, mudah diinterpretasi, dan cepat dalam pelatihan dan prediksi. Ini cocok untuk pemahaman awal tentang bagaimana variabel independen mempengaruhi variabel target.
   - **Catatan**: Namun, Logistic Regression bekerja dengan baik ketika hubungan antara variabel independen dan target adalah linier. Jika hubungannya kompleks, model ini mungkin tidak cukup kuat.

2. **Support Vector Machine (SVM)**:
   - **Alasan**: SVM adalah model yang sangat berguna ketika memiliki data yang tidak linier dan kompleks. Ini dapat menangani masalah klasifikasi biner dan dapat menangani permasalahan yang lebih rumit dengan transformasi kernel.
   - **Keuntungan**: SVM dapat mengidentifikasi batas keputusan yang kompleks, bahkan dalam kasus yang tidak linier. Hal ini dapat digunakan untuk masalah yang tidak dapat dipecahkan dengan baik oleh model linier seperti Logistic Regression.
   - **Catatan**: SVM dapat memerlukan penyetelan parameter yang hati-hati, dan waktu pelatihan yang lebih lama tergantung pada kompleksitas data.

3. **k-Nearest Neighbors (KNN)**:
   - **Alasan**: KNN adalah model yang berguna ketika ingin mempertimbangkan tetangga terdekat dari setiap titik data. Ini cocok untuk permasalahan klasifikasi dan dapat digunakan dalam konteks regresi. 
   - **Keuntungan**: KNN tidak membuat asumsi tentang bentuk fungsi keputusan seperti model linier. Ia dapat menangani data non-linier dan kompleks. Hal ini cocok untuk data yang memiliki pola tetangga yang penting.
   - **Catatan**: KNN dapat sensitif terhadap parameter seperti jumlah tetangga yang digunakan, dan ia memerlukan penyimpanan semua data pelatihan, yang dapat menjadi masalah pada dataset besar.

Pemilihan model harus didasarkan pada karakteristik data dan tujuan analisis. Idealnya, mungkin melakukan percobaan ketiga model dan melakukan validasi silang untuk menentukan model mana yang memberikan kinerja terbaik dalam konteks dataset.

## Hyperparameter Tuning


1. **Logistic Regression**:
   - **Hyperparameter**: `C` (kekuatan regulasi), `fit_intersep` dan `penalty` (jenis regulasi, l1 atau l2).
   - **Tuning dengan RandomizedSearchCV**: Dalam contoh di atas, kami menggunakan `RandomizedSearchCV` untuk mencari kombinasi optimal dari `C` dan `penalty` dengan melakukan pencarian acak di antara berbagai nilai `C` dan jenis `penalty`. Hasilnya adalah model Logistic Regression yang dioptimalkan.

2. **Support Vector Machine (SVM)**:
   - **Hyperparameter**: `C` (kekuatan regulasi) dan `kernel` (jenis kernel, linear atau rbf).
   - **Tuning dengan RandomizedSearchCV**: Untuk SVM, kami melakukan tuning parameter `C` dan `kernel` dengan `RandomizedSearchCV`. Ini membantu mencari kombinasi hyperparameter terbaik yang mengoptimalkan kinerja model SVM.

3. **k-Nearest Neighbors (KNN)**:
   - **Hyperparameter**: `n_neighbors` (jumlah tetangga) dan `metric` (jenis metrik jarak, seperti Euclidean atau Manhattan).
   - **Tuning dengan RandomizedSearchCV**: Untuk model KNN, kami menggunakan `RandomizedSearchCV` untuk menemukan nilai `n_neighbors` dan jenis `metric` yang paling cocok untuk dataset. Ini membantu memaksimalkan kinerja model KNN.

Dengan `RandomizedSearchCV`,  dapat melakukan eksplorasi acak parameter-parameter ini dalam jangka waktu yang lebih singkat daripada pencarian grid dan tetap mendapatkan model yang dioptimalkan.
